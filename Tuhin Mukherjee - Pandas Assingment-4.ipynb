```python
# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]
```

Ex- 

V Kohli:

Innings:  

Total Runs: 

No. of MOM(Man of the match):

No. of boundries:

```python
def record(name):
    mask=megaa[['batsman','inning','total_runs','batsman_runs','player_of_match']]
    mask1=mask[mask['batsman']==name]
    innings=mask1['inning'].count()
    total_runs=mask1['batsman_runs'].sum()
    mom=matches[(matches['player_of_match']==name)]['player_of_match'].count()
    boundries=mask1[(mask1['batsman_runs']==4) | (mask1['batsman_runs']==6)]['batsman_runs'].count()
    print("Batting Profile of :",name)
    print("innings :",innings,"\nTotal Runs :",total_runs,"\nNo. of MOM(Man of the match): ",mom,"\nNo. of boundries : ",boundries)
```

```python
name=input("Enter player name : \t")
record(name)
```

```python
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```python
def topper(vanue):
    mask=megaa[['batsman','batsman_runs','venue']]
    mask1=mask[mask['venue']==vanue].groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(4)
   
    print("The top-4 batsman at the given venue : \n",mask1)
```

```python
vanue=input("Enter the name of the venue : \t")
topper(vanue)
```

```python
vanue=input("Enter the name of the venue : \t")
topper(vanue)
```
