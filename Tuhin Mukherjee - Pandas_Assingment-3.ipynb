## Q3. Make a function to find top bowler(in terms of most wickets) against any team. Take input as team.

```python
deliveries.head()
```

```python
def top_b(team):
    full_name={
            "MI" : "Mumbai Indians",
            "CSK" : "Chennai Super Kings",            
            "KKR" : "Kolkata Knight Riders",          
            "RCB" : "Royal Challengers Bangalore",
            "KXIP" : "Kings XI Punjab",               
            "RR" : "Rajasthan Royals",               
            "DD" : "Delhi Daredevils",               
            "SRH" : "Sunrisers Hyderabad",            
            "DC" : "Deccan Chargers",                
            "GL" : "Gujarat Lions",                  
            "PW" : "Pune Warriors",                  
            "RPS" : "Rising Pune Supergiant",         
            "KTK" : "Kochi Tuskers Kerala",            
            "RPSS" : "Rising Pune Supergiants",         
            }
    op=""
    
    for word1 in team:
        op=full_name.get(team.upper(),team)
        
    mask=deliveries[(deliveries['batting_team']==op) & (deliveries['dismissal_kind'].isin(['caught','bowled','lbw','stumped','caught and bowled']))]
    mask1=mask['bowler'].value_counts().head(1)
    print(mask1)
```

```python
team=input("Enter team's name : \t")
top_b(team)
```

## Q4. Write a function which will take team name as input and return it's record of no. occurance in finals and no. of times won.  [Difficult]

ex.-- if team is CSK-- 
CSK in finals-

Played- 8
Won- 3

If any team which has not played in finals like Gujrat Lions

Gujrat Lions in finals-

Played-0


```python
def final_won(team):
    full_name={
            "MI" : "Mumbai Indians",
            "CSK" : "Chennai Super Kings",            
            "KKR" : "Kolkata Knight Riders",          
            "RCB" : "Royal Challengers Bangalore",
            "KXIP" : "Kings XI Punjab",               
            "RR" : "Rajasthan Royals",               
            "DD" : "Delhi Daredevils",               
            "SRH" : "Sunrisers Hyderabad",            
            "DC" : "Deccan Chargers",                
            "GL" : "Gujarat Lions",                  
            "PW" : "Pune Warriors",                  
            "RPS" : "Rising Pune Supergiant",         
            "KTK" : "Kochi Tuskers Kerala",            
            "RPSS" : "Rising Pune Supergiants",         
            }
    opp=""
    
    for word1 in team:
        opp=full_name.get(team.upper(),team)
        
    df=matches.drop_duplicates(subset=['season'],keep='last')
    mpinf=df[(df['team1']==opp) | (df['team2']==opp)]['season'].count()
    mwinf=df[df['winner']==opp]['season'].count()
    print(opp," played ",mpinf," matches in final & won ",mwinf," matches in final")
    
```

```python
team=input("Enter team's name :\t")
final_won(team)
```

## Q5. Write a function which will take a input as batsman and in output it gives total runs scored against each team. [Medium]




```python
def bat(batsman_name):
    mask1=deliveries[deliveries['batsman']==batsman_name]
    mask2=mask1[['bowling_team','batsman','total_runs']]
    mask3=mask1['bowling_team'].value_counts()
    print("The total score of ",batsman_name," aginst each team is as follows :\n")
    print(mask3)
```

```python
batsman=input("Enter the batsman name ")
bat(batsman)
```
