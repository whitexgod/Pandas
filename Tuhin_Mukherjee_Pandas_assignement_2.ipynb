```python
# Q3 : Write a function which will take 2 inputs, team 1 and team 2 and their record against each other

# For example if team1->KKR and team2->CSK

# The output should be like KKR 2(matches won by KKR) and CSK 5(matches won by CSK)[Medium]
```

```python
def compare(team1,team2):
    full_name={
            "MI" : "Mumbai Indians",
            "CSK" : "Chennai Super Kings",            
            "KKR" : "Kolkata Knight Riders",          
            "RCB" : "Royal Challengers Bangalore",
            "KXIP" : "Kings XI Punjab",               
            "RR" : "Rajasthan Royals",               
            "DD" : "Delhi Daredevils",               
            "SH" : "Sunrisers Hyderabad",            
            "DC" : "Deccan Chargers",                
            "GL" : "Gujarat Lions",                  
            "PW" : "Pune Warriors",                  
            "RPS" : "Rising Pune Supergiant",         
            "KTK" : "Kochi Tuskers Kerala",            
            "RPSS" : "Rising Pune Supergiants",         
            }
    output1=""
    output2=""
    for word1 in team1:
        output1=full_name.get(team1.upper(),team1)
    for word2 in team2:
        output2=full_name.get(team2.upper(),team2)
    
    result=df[( (df['team1']==output1) & (df['team2']==output2) ) | ( (df['team1']==output2) & (df['team2']==output1) )]
    
    mask1=result[result['winner']==output1].shape[0]
    mask2=result[result['winner']==output2].shape[0]
    print(team1,'won',mask1,'matchs','and',team2,'won',mask2,'matches') 
```

```python
team1=input('team 1=')
team2=input('team 2=')
compare(team1,team2)
```

```python
def check(team):
    full_name={
            "MI" : "Mumbai Indians",
            "CSK" : "Chennai Super Kings",            
            "KKR" : "Kolkata Knight Riders",          
            "RCB" : "Royal Challengers Bangalore",
            "KXIP" : "Kings XI Punjab",               
            "RR" : "Rajasthan Royals",               
            "DD" : "Delhi Daredevils",               
            "SH" : "Sunrisers Hyderabad",            
            "DC" : "Deccan Chargers",                
            "GL" : "Gujarat Lions",                  
            "PW" : "Pune Warriors",                  
            "RPS" : "Rising Pune Supergiant",         
            "KTK" : "Kochi Tuskers Kerala",            
            "RPSS" : "Rising Pune Supergiants",         
            }
    output1=""
    
    for word1 in team:
        output1=full_name.get(team.upper(),team)
    
    toss_winning=df[(df['toss_winner']==output1)].shape[0]
    win_after_toss_win=df[(df['toss_winner']==output1) & (df['winner']==output1)].shape[0]
    percentage=(win_after_toss_win/toss_winning)*100
    print('The winning percentage of the team after winning the toss = ',percentage)
```
